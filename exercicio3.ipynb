2. Crie uma função que calcule e imprima velocidade media de um objeto a partir de uma posição inicial, a final e o tempo transcorrido para um objeto em MRU. Também crie uma funcão que calcule e imprima a velocidade de um objeto a partir da aceleração constante e o tempo (MRUA) (p.ex. queda libre).


```python
def calculo_velocidade_media(distancia_inicial,distancia_final,tempo_decorrido):
  velocidade_media = (distancia_final-distancia_inicial)/tempo_decorrido
  print("A velocidade média é:",velocidade_media)

def calculo_velocidade(aceleracao,tempo_decorrido):
  velocidade = aceleracao*tempo_decorrido
  print("A velocidade final é:",velocidade)

calculo_velocidade_media(0,10,2)
calculo_velocidade(10,2)
```

4. Crie uma função que faça a conversão de uma medida inicialmente em milhas para m, e outra para o inverso; uma de horas para segundos, e o inverso. Utilize estas funções para resolver novamente o primeiro exercício da semana passada (da corrida). Se uma pessoa demora 30 minutos em 4 milhas, qual velocidade media em km/h ? e o tempo medio por kilometro?


```python
def milha_to_quilometro(milha):
  quilometro = milha*1.61
  return quilometro

def quilometro_to_milha(quilometro):
  milha = quilometro/1.61
  return milha

def hora_to_segundo(hora):
  segundo = hora*3600
  return segundo

def minuto_to_segundo(minuto):
  segundo = minuto*60
  return segundo

def segundo_to_minuto(segundo):
  minuto = segundo/60
  return minuto

def segundo_to_hora(segundo):
  hora = segundo/3600
  return hora
```

```python
# Se você fizer uma corrida de 10 quilômetros em 43 minutos e 30 segundos, qual será seu tempo médio por milha? 
# Qual é a sua velocidade média em milhas por hora? (Dica: há 1,61 quilômetros em uma milha).

tempo_segundo = minuto_to_segundo(43) +  30
tempo_hora = segundo_to_hora(tempo_segundo)
distancia_milha = quilometro_to_milha(10)

tempo_medio =  tempo_segundo/distancia_milha # segundo/milha
velocidade_media = distancia_milha/tempo_hora # milha/hora

minutos = segundo_to_minuto(tempo_medio)

print("O tempo médio por milha é:",minutos,"minutos")
print("A velocidade média em milhas por hora é:",velocidade_media,"mph")
```

```python
# Se uma pessoa demora 30 minutos em 4 milhas, qual velocidade media em km/h ? e o tempo medio por kilometro?

tempo_segundo = minuto_to_segundo(30)
tempo_hora = segundo_to_hora(tempo_segundo)
distancia_quilometro = milha_to_quilometro(4)

tempo_medio =  tempo_segundo/distancia_quilometro # segundo/quilometro
velocidade_media = distancia_quilometro/tempo_hora # quilometro/hora

minutos = segundo_to_minuto(tempo_medio)

print("O tempo médio por milha é:",minutos,"minutos")
print("A velocidade média em milhas por hora é:",velocidade_media,"km/h")
```
