```python
# Questão 1
velocidade_1 = 12 # km/h
distancia_1 = 5 # km
intervalo = 1/60 # h

distancias = []
distancia = velocidade_1*intervalo
while distancia <= distancia_1:
  distancias.append(round(distancia,4))
  distancia += velocidade_1*intervalo

# Questão 2
distancia += 0.2
velocidade_2 = 15 # km/h
distancia_2 = 7 # km
while distancia <= distancia_2:
  distancias.append(round(distancia,4))
  distancia += velocidade_2*intervalo

import matplotlib.pyplot as plt

t = [i for i in range(1,len(distancias)+1)]
x = distancias

plt.plot(t,x)
plt.xlabel("Tempo (min)")
plt.ylabel("Distância (km")

print("O percurso durou um pouco mais que",t[-1],"minutos")
```

```python
# Questão 3
import math
import numpy as np

def function_teta(t,teta_zero,g,l):
  zeta = (g/l)**(1/2)*t
  return teta_zero*math.cos(zeta)

teta_zero = math.radians(90)
g = 9.8 # m/s²
l = 1 # m

xs = []
ys = []
tetas = []
for t in np.arange(0,10,0.01):
  teta = function_teta(t,teta_zero,g,l)
  tetas.append(teta)
  x = l*math.sin(teta)
  xs.append(x)
  y = l*math.cos(teta)
  ys.append(y)

t = [i for i in np.arange(0,10,0.01)]

plt.plot(t,xs)
plt.xlabel("Tempo (s)")
plt.ylabel("x (m)")
plt.show()

plt.plot(t,ys)
plt.xlabel("Tempo (s)")
plt.ylabel("y (m)")
plt.show()
```
